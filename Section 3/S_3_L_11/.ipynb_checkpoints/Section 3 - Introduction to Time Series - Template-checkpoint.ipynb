### Import the Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

### Importing the Data

```python
raw_csv_data = pd.read_csv('Index2018.csv')
```

```python
df_comp = raw_csv_data.copy()
```

### Plotting the Data

```python
df_comp.spx.plot(figsize=(20,5), title='SPX')
```

```python
df_comp.dax.plot(figsize=(20,5), title='DAX')
```

```python
df_comp.spx.plot(figsize=(20,5), title='SPX')
df_comp.dax.plot(figsize=(20,5), title='DAX')
plt.title('SPX vs DAX')
```

### The QQ Plot

```python
from scipy.stats import probplot
import pylab
```

```python
probplot(df_comp.spx, plot=pylab)
plt.title('QQ Plot', size=24)
pylab.show()
```
